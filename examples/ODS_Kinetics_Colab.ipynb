```python
# ── Cell 2: Generate Excel Template ──
from catlab.ods_kinetics import generate_template

catalyst_names = [
    'Catalyst_1',   # <- replace with your catalyst names
    'Catalyst_2',
    'Catalyst_3',
]

generate_template(
    output_path='ods_template.xlsx',
    example_catalysts=catalyst_names
)

from google.colab import files
files.download('ods_template.xlsx')
print('Template downloaded. Fill it and upload in Cell 3.')
```

```python
# ── Cell 3: Upload & Analyse ──
from google.colab import files
from catlab.ods_kinetics import run_ods_analysis
from IPython.display import Image, display
import os

print('Upload your filled Excel file:')
uploaded = files.upload()
excel_path = list(uploaded.keys())[0]

# ════ SET YOUR PARAMETERS HERE ════
results = run_ods_analysis(
    excel_path      = excel_path,
    c0_value        = 250.0,        # <- initial concentration
    c0_unit         = 'ppmS',       # <- ppmS | ppm | mg/L | g/L | mmol/L | mol/L
    catalyst_mass_g = 0.05,         # <- catalyst mass (g)
    solution_vol_L  = 0.010,        # <- fuel volume (L)
    mw_pollutant    = None,         # <- molecular weight (only for ppm/mg/L/g/L)
    output_dir      = '/content/ods_output',
)

# Display plots
for fname in ['01_conversion_vs_time.png', '02_zero_order.png',
              '03_pseudo_first_order.png', '04_second_order.png']:
    p = f'/content/ods_output/{fname}'
    if os.path.exists(p):
        display(Image(p))

# Download all outputs
for f in os.listdir('/content/ods_output'):
    files.download(f'/content/ods_output/{f}')
```
